```
from tkinter import *
from tkinter import messagebox as ms
import sqlite3

# Kullanıcılar ve program başlangıcında veritabanları
with sqlite3.connect('çıkış.db') as db:
    c = db.cursor()

c.execute('Kullanıcı adı mevcut degılse tablo olustur (Metin boşsa isim anahtarı,metin boşsa sifre );')
db.commit()
db.close()

#AnaBölüm
class main:
    def __init__(self,master):
    	# Pencere 
        self.master = master
        # KullanışlıDegiskenler
        self.username = StringVar()
        self.password = StringVar()
        self.n_username = StringVar()
        self.n_password = StringVar()
        #UzantıYarat
        self.widgets()

    #Login Function
    def login(self):
    	#Establish Connection
        with sqlite3.connect('çıkış.db') as db:
            c = db.cursor()

        #Find user If there is any take proper action
        find_user = ('Kullanıcı Adı İçin(kullanıcı adı,sıfre) DEĞERLER(?,?)')
        c.execute(find_user,[(self.username.get()),(self.password.get())])
        result = c.fetchall()
        if result:
            self.logf.pack_forget()
            self.head['metin'] = self.username.get() + '\n Gırıs'
            self.head['boyut'] = 150
        else:
            ms.showerror('Maalesef!','Kullanıcı Bulunamadı.')
            
    def new_user(self):
    	#BağlantıKurmak
        with sqlite3.connect('cıkıs.db') as db:
            c = db.cursor()

        #İşlem için kullanıcı adını bulunuz
        find_user = ('Kullanıcı adı nerede ve buton = ?')
        c.execute(find_user,[(self.n_username.get())])        
        if c.fetchall():
            ms.showerror('Hata!','Kullanıcı Adı alındı ve farklı birini deneyin.')
        else:
            ms.showinfo('Başarılı!','Kabul Edildi!')
            self.log()
        #Yeni Hesap Oluşturma 
        insert = 'Kullanıcı Adı İçin(kullanıcı adı,sıfre) DEĞERLER(?,?)'
        c.execute(insert,[(self.n_username.get()),(self.n_password.get())])
        db.commit()

        # Paket
    def log(self):
        self.username.set('')
        self.password.set('')
        self.crf.pack_forget()
        self.head['metin'] = 'GIRIS'
        self.logf.pack()
    def cr(self):
        self.n_username.set('')
        self.n_password.set('')
        self.logf.pack_forget()
        self.head['metın'] = 'Yenı Hesap Olustur'
        self.crf.pack()
        
    #TasarımCiz
    def widgets(self):
        self.head = Label(self.master,text = 'GIRIS',font = ('',35),pady = 10)
        self.head.pack()
        self.logf = Frame(self.master,padx =10,pady = 10)
        Label(self.logf,text = 'KullanıcıAdı: ',font = ('',20),pady=5,padx=5).grid(sticky = W)
        Entry(self.logf,textvariable = self.username,bd = 5,font = ('',15)).grid(row=0,column=1)
        Label(self.logf,text = 'Sıfre: ',font = ('',20),pady=5,padx=5).grid(sticky = W)
        Entry(self.logf,textvariable = self.password,bd = 5,font = ('',15),show = '*').grid(row=1,column=1)
        Button(self.logf,text = ' Gırıs ',bd = 3 ,font = ('',15),padx=5,pady=5,command=self.login).grid()
        Button(self.logf,text = ' HesapYarat ',bd = 3 ,font = ('',15),padx=5,pady=5,command=self.cr).grid(row=2,column=1)
        self.logf.pack()
        
        self.crf = Frame(self.master,padx =10,pady = 10)
        Label(self.crf,text = 'KullanıcıAdı: ',font = ('',20),pady=5,padx=5).grid(sticky = W)
        Entry(self.crf,textvariable = self.n_username,bd = 5,font = ('',15)).grid(row=0,column=1)
        Label(self.crf,text = 'Ismı: ',font = ('',20),pady=5,padx=5).grid(sticky = W)
        Entry(self.crf,textvariable = self.n_password,bd = 5,font = ('',15),show = '*').grid(row=1,column=1)
        Button(self.crf,text = 'KullanıcıYarat',bd = 3 ,font = ('',15),padx=5,pady=5,command=self.new_user).grid()
        Button(self.crf,text = 'GIR',bd = 3 ,font = ('',15),padx=5,pady=5,command=self.log).grid(row=2,column=1)

        if __name__ == '__AnaDurum__':
	
	
         root = Tk()
         root.title('GırısYap')
    #geometriYukle('400x350+300+300')
         main(root)
         root.mainloop()
```
