```

from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing import image
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
from keras.utils import to_categorical
print('Veriler : ', train_images.shape, train_labels.shape)
print('Tahmini veriler : ', test_images.shape, test_labels.shape)
sınıf = np.unique(train_labels)
nSınıf = len(sınıf)
print('Toplam rakam : ', nSınıf)
print('Sonuç : ', sınıf)

plt.figure(figsize=[10,5])


plt.figure(figsize=(10,10))
for i in range(16):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(test_images[i, :, :], cmap='gray')
    plt.title("Gerçek sonuç : {}".format(test_labels[i]))
plt.show()
```
